## 2a. $|in(n-m)|^{(p-1)}$

### $|in(n-m)|$:
\begin{gather}
in(1) = a_1 + jb_1\\

in = 

\begin{bmatrix}
a_1 & b_1 \\
a_2 & b_2 \\
a_3 & b_3 \\
\vdots & \vdots \\
a_n & b_n
\end{bmatrix}

\\

in*in' = 

\begin{bmatrix}
a_1 & b_1 \\
a_2 & b_2 \\
a_3 & b_3 \\
\vdots & \vdots \\
a_n & b_n
\end{bmatrix}

*

\begin{bmatrix}
a_1 & a_2 & a_3 & \dots & a_n \\
b_1 & b_2 & b_3 & \dots & b_n \\
\end{bmatrix}
=\\

in*in' = 

\begin{bmatrix}
a_1^2 + b_1^2 \\
a_2^2 + b_2^2 \\
a_3^2 + b_3^2 \\
\vdots \\
a_n^2 + b_n^2
\end{bmatrix}

\end{gather}

Tirar a raiz quadrada de cada elemento:

\begin{gather}

\begin{bmatrix}
\sqrt{a_1^2 + b_1^2} \\
\sqrt{a_2^2 + b_2^2} \\
\sqrt{a_3^2 + b_3^2} \\
\vdots \\
\sqrt{a_n^2 + b_n^2}
\end{bmatrix}

= |in|

\end{gather}

Deslocar os valores da matriz para baixo dependendo do valor de m, por exemplo se m=1:

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2} \\
\sqrt{a_2^2 + b_2^2} \\
\vdots \\
\sqrt{a_n^2 + b_n^2}
\end{bmatrix}


### $|in(n-m)|^{(p-1)}$

#### p=1:

Quando p=1 a matriz resultante vai ser composta de números 1 (não é necessário realizar operações):

\begin{bmatrix}
1 \\
1 \\
1 \\
\vdots \\
1
\end{bmatrix}

#### p=2:

Quando p=2 a matriz resultante vai ser igual a in(n-m) (não é necessário realizar operações): 

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2} \\
\sqrt{a_2^2 + b_2^2} \\
\vdots \\
\sqrt{a_n^2 + b_n^2}
\end{bmatrix}

(p/ m=1)

Até então temos: 

\begin{bmatrix}
1 & 0 \\
1 & \sqrt{a_1^2 + b_1^2} \\
1 & \sqrt{a_2^2 + b_2^2} \\
\vdots & \vdots \\
1 & \sqrt{a_n^2 + b_n^2}
\end{bmatrix}

#### p=n com n>2:

Para as próximas etapas com p>2 basta pegar a última coluna calculada e  multiplicar pela transposta de in(n-m).
Para p=3 temos:

\begin{gather}

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2} \\
\sqrt{a_2^2 + b_2^2} \\
\vdots \\
\sqrt{a_n^2 + b_n^2}
\end{bmatrix}
*
\begin{bmatrix}
0 & \sqrt{a_1^2 + b_1^2} & \sqrt{a_2^2 + b_2^2} & \dots & \sqrt{a_n^2 + b_n^2}
\end{bmatrix}

=\\

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2}^2 \\
\sqrt{a_2^2 + b_2^2}^2 \\
\vdots \\
\sqrt{a_n^2 + b_n^2}^2
\end{bmatrix}
\end{gather}

A matriz até agora: 

\begin{bmatrix}
1 & 0 & 0 \\
1 & \sqrt{a_1^2 + b_1^2} & \sqrt{a_1^2 + b_1^2}^2 \\
1 & \sqrt{a_2^2 + b_2^2} & \sqrt{a_2^2 + b_2^2}^2 \\
\vdots & \vdots & \vdots \\
1 & \sqrt{a_n^2 + b_n^2} & \sqrt{a_n^2 + b_n^2}^2
\end{bmatrix}

Para p=4 temos:

\begin{gather}

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2}^2 \\
\sqrt{a_2^2 + b_2^2}^2 \\
\vdots \\
\sqrt{a_n^2 + b_n^2}^2
\end{bmatrix}
*
\begin{bmatrix}
0 & \sqrt{a_1^2 + b_1^2} & \sqrt{a_2^2 + b_2^2} & \dots & \sqrt{a_n^2 + b_n^2}
\end{bmatrix}

=\\

\begin{bmatrix}
0 \\
\sqrt{a_1^2 + b_1^2}^3 \\
\sqrt{a_2^2 + b_2^2}^3 \\
\vdots \\
\sqrt{a_n^2 + b_n^2}^3
\end{bmatrix}
\end{gather}
